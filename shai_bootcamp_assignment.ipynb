```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Load your dataset
df = pd.read_csv("/content/drive/MyDrive/Salaries(exercise) (1).csv")
df.head()


df.columns
```

```python

### Basic Data Exploration:


# Identify the number of rows and columns
num_rows, num_columns = df.shape
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")

# Determine the data types of each column
data_types = df.dtypes
print("\nData types of each column:")
print(data_types)

# Check for missing values in each column
missing_values = df.isnull().sum()
print("\nMissing values in each column:")
print(missing_values)


### Descriptive Statistics:


# Calculate basic statistics
mean_salary = df['TotalPay'].mean()
median_salary = df['TotalPay'].median()
mode_salary = df['TotalPay'].mode()[0]
min_salary = df['TotalPay'].min()
max_salary = df['TotalPay'].max()
salary_range = max_salary - min_salary
std_salary = df['TotalPay'].std()

print("\nBasic Statistics:")
print(f"Mean Salary: {mean_salary}")
print(f"Median Salary: {median_salary}")
print(f"Mode Salary: {mode_salary}")
print(f"Minimum Salary: {min_salary}")
print(f"Maximum Salary: {max_salary}")
print(f"Salary Range: {salary_range}")
print(f"Standard Deviation of Salaries: {std_salary}")


### Data Cleaning:

#Let's fill missing values in 'TotalPay' with the median, a robust measure of central tendency:


# Fill missing values in 'TotalPay' with the median
df['TotalPay'].fillna(df['TotalPay'].median(), inplace=True)


### Basic Data Visualization:



plt.figure(figsize=(10, 6))
plt.hist(df['TotalPay'], bins=20, edgecolor='black', alpha=0.7)
plt.title('Distribution of Salaries')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')
plt.show()

# Use a bar chart to represent the top job titles
top_job_titles = df['JobTitle'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_job_titles.plot(kind='bar')
plt.title('Top 10 Job Titles')
plt.xlabel('Job Title')
plt.ylabel('Count')
plt.show()


### Grouped Analysis:

# Group the data by department and calculate average salaries
average_salary_by_department = df.groupby('JobTitle')['TotalPay'].mean()

# Compare the average salaries across different departments
print("\nAverage Salaries by Job Title:")
print(average_salary_by_department)


### Simple Correlation Analysis:


# Identify any correlation between salary and another numerical column (e.g., 'BasePay')
correlation = df['TotalPay'].corr(df['BasePay'])

# Plot a scatter plot to visualize the relationship
plt.figure(figsize=(10, 6))
plt.scatter(df['BasePay'], df['TotalPay'])
plt.title('Scatter Plot of TotalPay vs. BasePay')
plt.xlabel('Base Pay')
plt.ylabel('Total Pay')
plt.show()

print("\nCorrelation between TotalPay and BasePay:", correlation)

```
